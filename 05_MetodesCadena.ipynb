2.1 MESURES DE LES CADENES EN PYTHON:
len(cadena) Longitut d'una cadena
max(cadena) Valor més gran d'una cadena  
min(cadena) Valor més petit d'una cadena
ord(caràcter) Dóna un número
chr(número) Dóna un caràcter

```python
cadena = input("Llegueix el que escric: ")

#imprimir la longitut del que escric
print(len(cadena))

#imprimir el valor màxim
print(max(cadena))

#imprimir el valor mínim
print(min(cadena))

print(ord(cadena[0]))

print(chr(ord(cadena[6])))


```

2.3 CERCA DE CARACTERS A CADENES EN PYTHON: Mètodes per trobar caracters i posicions dins d'una cadena:count, index, rindex, find, rfind

```python
cadenacompta = "Comptar les lletres que conte aquesta frase"
#.count compte les lletres
print(cadenacompta.count('a'))

#les majúscules són diferents de les minúscules
print(cadenacompta.count('A'))

#index troba quina posició és la primera
print(cadenacompta.index('a'))

#index si no troba el caràcter dona un error, per tant poderm fer servir la funció find
#print(cadenacompta.index('A'))

print(cadenacompta.find('A'))

#Les funcions rindex o rfind, ho fan reverses, per que lleguix des de el final al començament

print(cadenacompta.rindex('a'))
print(cadenacompta.rfind('A'))
```

```python
# com modificar les cadenes en majúscules o minúscules, fent servir capitalize(), lower() i upper()
cadenaorigen = "poDrem, modiFicar en Funció de les necessitats."

print('1', cadenaorigen)

print('2', cadenaorigen.capitalize())

print('3', cadenaorigen.upper())

print('4', cadenaorigen.lower())

print('5', cadenaorigen.capitalize())


tiposletras = input('Introdueix una paraula i parlarem del case en les lletres')

#Modo marrano
#print('La cadena', tiposletras, 'en majúsculas és: ', tiposletras.upper(), 'en minúsculas és:', tiposletras.lower(), 'i amb la primera capital és:', tiposletras.capitalize(), 'bye, bye piccolino')

#Modo elegant
print(f"La cadena \t\t {tiposletras} \nen majúsculas és: \t {tiposletras.upper()} \nen minúsculas és: \t {tiposletras.lower()} \nprimera capital és: \t {tiposletras.capitalize()} \nbye, bye piccolino")
```


2.6 DIVISIÓ DE CADENES A PYTHON: Divisió de Cadenes a Python. Mètodes split(), partition(),rpartition().

```python

comentari = "gent hiperconcentrada-classe escoltant videos"

print(comentari.split(' '))

print(comentari.partition('-'))
print(type(comentari.partition('-')))
print(comentari.rpartition(' '))
```


2.7 ORDENACIÓ DELS CARACTERS D'UNA CADENA A PYTHON: Ordenació dels caracters d'una cadena a Python amb métode sorted i reverse, ordena segons el codi decimal d'un caràcter ASCII or Unicode

```python
pcadena = "A4z1b598"
print(pcadena)
print(sorted(pcadena))

pcadena_o1 = sorted(pcadena)
print(pcadena_o1)
print(''.join(pcadena_o1))


pcadena_o2 = sorted(pcadena, reverse = True)
print(pcadena_o2)
print(''.join(pcadena_o2))


```
