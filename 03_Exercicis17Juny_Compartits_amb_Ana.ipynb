69.	Com es fa per copiar tota una llista amb slicing? Fent una assignació amb una altre llista, on es fa un còpia superficial de la original, es a dir les modificacions posteriors no quedaran reflexades en la nova còpia

```python
'''help(slice.__doc__)'''
LlistaOriginal = [1, 2, [3, 4],5,[6,7]]
LlistaCopia = LlistaOriginal[:]  # Còpia superficial fent servir la funcio slicing

# Si es modifica la llista original, no es canvia la còpia de la llista
LlistaOriginal[0] = 10
print(f"Llista original: {LlistaOriginal}")
print(f"Llista copiada: {LlistaCopia}")

# Modificació d'un element de la subllista
LlistaOriginal[2][0] = 8
print(f"Llista original: {LlistaOriginal}")
print(f"Llista copiada: {LlistaCopia}")
```
