Exercici 1: Validació de correus electrònics
Objectiu: Escriu una expressió regular per validar correus electrònics simples.
Entrada: hola@test.com, josep.puigdemont@company.cat, marta123@xyz.org, persona@, ert@rtt.trt.rt,


````python
import re

cadena ='''hola@test.com, hola@test..com., hola@test.com.., josep.puigdemont@company.cat, marta123@xyz.org, persona@., trewt@rtt.dfs.ar'''

patro = re.compile("[a-z0-9]+@[a-z]+(?:\.[a-z]+){1,}") # no detecta que 'hola@test.com..' està mal escrit

resultat = patro.findall(cadena)
print(resultat)
````

Exercici 5: Extracció de preus en format decimal
Objectiu: Escriu un regex per trobar preus amb decimals en una cadena de text.
Entrada: Els preus dels productes són 10.99€, 20€, 100.50€, i 300.00€.
OjO Look ahead & behind


````python
import re

cadena ="10.99€, 20€, 100.50€, i 300.00€, .50€, 49.€"

patro = re.compile("[0-9]*\.[0-9]*€") # en aquest cas acepta .50€, 49.€
# patro = re.compile("[0-9]*\.[0-9]+€") -- Obliga que hi hagin digits decimals
# patro = re.compile("[0-9]*\.{0,1}[0-9]+€") aquest patró agafa també preus sense decimals

resultat = patro.findall(cadena)
print(resultat)
````

Exercici 6: Validació de dates en format DD/MM/AAAA
Objectiu: Escriu un regex per validar dates en format DD/MM/AAAA.
Entrada: 25/12/2021, 01/01/1999, 31/02/2020, 30/04/2020.


````python
import re

cadenafecha ="25/12/2021, 01/01/1999, 31/02/2020, 30/04/2020, 45/12/2020, 01/24/2020, 01/04/2050"

patro = re.compile("(?:[012][0-9]|30|31)/(?:0[1-9]|1[0-2])/(?:19[0-9]{2}|20[01][0-9]]|202[0-5])")

fechas = patro.findall(cadenafecha)
print(fechas)
# el resultat amb colab google no és l'esperat, està comprovat amb pythex i funciona bé el patró -> SI funciona correctament
````

Exercici 11: Validació de contrasenyes
Objectiu: Escriu un regex per validar contrasenyes que tinguin entre 8 i 16 caràcters,
almenys una majúscula, una minúscula, i un número.
Entrada: Contraseña1, pass, PASSWORD123, qwert7op, Pp34567,12345678 , qwertyui, QWERTYUI, ppppppp8, 12345678P, PALpal123123PALpal


**Aquesta versió funciona **

JOAN la teva proposta de look ahead no funciona


```
^(?=[a-zA-Z0-9]{8,16}$)(?=.*[A-Z])(?=.*[a-z])(?=.*\d)[a-zA-Z0-9]+$
```



````python
import re
cadena ="Contraseña1,pass,PASSWORD123,qwert7op,Pp34567,12345678,qwertyui,QWERTYUI,ppppppp8,12345678P,PALpal123123PALpa, QpQp56789, 1pwerSD8"
patro = re.compile("(?=[a-zA-Z0-9]{8,16}$)(?=.*[A-Z])(?=.*[a-z])(?=.*\d)[a-zA-Z0-9]+")
resultat = patro.findall(cadena)
print(resultat)
````

````python
# ? 0 o 1 vez
# ?: se utiliza para crear un grupo de no capturar. Grupos de captura (), Grupos de NO captura (?:)
# (?:abc){2}  Busca 'abcabc'
# ?= sirve para verificar una condición en la cadena sin incluirla en la coincidencia final
# (?=.*[A-Z]) Assegura que hi ha almenys una majúscula després del punt actual
# \W coincide con cualquier caracter que no sea una palabra
#
import re

cadena ="Contraseña1,pass,PASSWORD123,qwert7op,Pp34567,12345678,qwertyui,QWERTYUI,ppppppp8,12345678P,PALpal123123PALpa, QpQp56789, 1pwerSD8"
llistapassword = cadena.split(",")
#print(llistapassword)
llistapatro = []
resultat = []

patro = re.compile("(?=.*[a-zñ])(?=.*[A-ZÑ])(?=.*[0-9])[A-ZÑa-zñ0-9]{8,16}")

for password in llistapassword:
    if len(password) >= 8 and len(password) <= 16:
        resultat.append(patro.findall(password))
        llistapatro.append(password)

#print(llistapatro)
print(resultat)


#patro = re.compile("(?=.+[a-zñ])(?=.+[A-ZÑ])(?=.+\d)[A-ZÑa-zñ\d]{8,16}")
#patro = re.compile((?=.*[a-zñ])(?=.*[A-ZÑ])(?=.*\d)[A-ZÑa-zñ\d]{8,16})
#patro = re.compile("[A-ZÑa-zñ\d]{8,16}")
# patro = re.compile("(?=.*[a-zñ])(?=.*[A-ZÑ])(?=.*[0-9])[A-ZÑa-zñ0-9]{8,16}")
#patro = re.compile("(?=[^a-zñ]*[a-zñ])(?=[^A-ZÑ]*[A-ZÑ])(?=[^\d]*\d)[A-ZÑa-zñ\d]{8,16})





````

Objectiu: Escriu un regex per validar horaris en format HH:MM (de 24 hores).
Entrada: 14:30, 02:59, 25:00.


````python
import re

cadenahora ="14:30, 02:59, 25:00, 22:70, 12:69"

patro = re.compile("(?:[01][0-9]|2[0-4])\:(?:[0-5][0-9])") # JOAN Corregit

horas = patro.findall(cadenahora)
print(horas)
````
