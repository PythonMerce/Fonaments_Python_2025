1- Invertir el valor d'una cadena:

a) Assignar a una variable una cadena de text qualsevol, després crear una
altra variable de text buida on haurem de fer el procés per passar els valors de e l'anterior cadena però invertida.
b) Mostrar els caràcters d'una cadena en ordre invers

```python
#range star, step, stop
#Metode A
cadena1 = "Cuando la noche te confunde"
cadena2= ""

for posicio in range(len(cadena1) -1, -1, -1):
    cadena2 += cadena1[posicio]
print('cadena original:', cadena1)
print('cadena invertida:', cadena2)

# Metode B
cadena1 = "Cuando la noche te confunde"
cadena2 = cadena1[::-1]
print(cadena2)

```

```python
#b) Mostrar els caràcters d'una cadena en ordre invers
cadena1 = "Cuando la noche te confunde"

for posicio in range(len(cadena1) -1, -1, -1):
    print(cadena1[posicio])
```

12- Escriu un programa que rebi 2 cadenes i ens digui si la primera cadena està dins de l'altra.
'el' 'la casa no estaba' --> Treu per pantalla: La cadena 'el' NO està dins de 'la casa no estaba'
'el' 'la casa del campo' --> Treu per pantalla: La cadena 'el' SI està dins de 'la casa del campo'

Operador	Descripción	Ejemplo	Valor de retorno
==	Las cadenas son iguales	"Hello" == "hello"	false
!=	Las cadenas no son iguales	"Hello" != "hello"	true
<	orden lexicográfico menor	"Australia" < "Zimbabwe"	true
>	orden lexicográfico mayor	"Australia" > "Zimbabwe"	false
<=	orden lexicográfico menor o igual que	"Germany" <= "United States"	false
>=	orden lexicográfico mayor o igual que	"Germany" <= "Germany"	true


```python

frase1 = str.lower("la casa no estaba")
frase2 = str.lower("la ")

if frase2 == frase1:
    print("la cadena ", frase2, 'SI está en:', frase1)

else:
    print("la cadena ", frase2, 'NO está en:', frase1)
```

1- Dissenya un programa que tregui tots els prefixes d'una cadena que rebem per paràmetre.
Per exemple la cadena "Fedora" treu per pantalla:
F
Fe
Fed
Fedo

Fedora

```python
cadena = input('Introduex una cadena:')
cadenatemp = []

#Sense Slicing

for posicio in range(len(cadena)):
    cadenatemp.append(cadena[posicio])
    print("".join(cadenatemp))

# Amb Slicing

cadenaslic = cadena[:]
for posicio in range(len(cadena)):
    print(cadenaslic[0:posicio+1])
```

2- Dissenya un programa que rebi per paràmetre una cadena i tregui per pantalla totes les subcadenes de longitud 3. Per exemple la cadena "planes" treu per pantalla:
pla

nes

```python
cadena = input('Introduex una cadena:')
cadenatemp = cadena[:]

for posicio in range(len(cadena)):
    if (len(cadena) - posicio ) > 3:
        print(cadenatemp[posicio:posicio+3])
```
