6. Sumar els múltiples de 2 menors a 100

**7.  Sumar les potències de 2 menors a 100**

**8. Calcular la mitjana dels números positius llegits**

```python
llistanumeros = [1, 3, 1, 3, 1, 3, 1, 3, 2 ,4]
divisor = len(llistanumeros)
index = 0
numerador = 0

while index < divisor:
   numerador = numerador + llistanumeros[index]
   index += 1

mitjana = numerador / divisor

print(f"La mitjana aritmètica del números {llistanumeros} és: {mitjana}")

```

**9. Dir si un número és primer o no**

```python
# Els números primers només són divisora amb 1 i amb ell mateix
numero = input("Entra un número per comprovar si és un número primer: ")
divisor = 2
llistadivisibles = []

if numero.isdigit():
    while (divisor < int(numero)):
        if int(numero) % divisor == 0:
            llistadivisibles.append(divisor)
        divisor +=1

    if len(llistadivisibles)> 0:
        print(f"{int(numero)} NO es número primer pq és divisible per: {llistadivisibles}")
    else:
        print(f"{int(numero)} és número PRIMER")
else:
    print(f"{(numero)} NO és un número, conté carácters no vàlids")



```
