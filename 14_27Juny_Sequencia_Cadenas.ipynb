
**División de cadenes**
```
split ----- cadena.split(caràcter)

```

OjO, en un CSV al final d'una línea hi ha final del carro \n



````python
cadena1 = "'New Yersey';'NY';'NY Police Deparment';'34';'67';'45';'32'\n"
print(cadena1, type(cadena1))

cadena2 = "New Yersey;NY;NY Police Deparment;34;67;45;32\n"
print(cadena2, type(cadena2))

linea1 =  cadena1.split(';')
print(linea1, type(linea1))

linea2 =  cadena2.split(';')
print(linea2, type(linea2))

linea12 =  cadena1.partition(';')
print(linea12, type(linea12))

linea22 =  cadena2.partition(';')
print(linea22, type(linea22))

linea13 =  cadena1.rpartition(';')
print(linea13, type(linea13))

linea23 =  cadena2.rpartition(';')
print(linea23, type(linea23))

````
