**2. Funció que pren com a paràmetre una llista de números i retorna la suma dels seus elements.**

```
def suma([4,3,2,1,8]) ---------> OUTPUT: 18
def suma([5,4,3,2,1]) ---------> OUTPUT: 15
```


````python
llista1 = [4, 3, 2, 1, 8]
llista2 = [5, 4, 3, 2, 1]

def sumallistes(llistasumar):
    resultat = 0
    for pos in range(len(llistasumar)):
        resultat = resultat + llistasumar[pos]
    return resultat

print(f"La suma del components de la llista: {llista1} es igual = {sumallistes(llista1)}")
print(f"La suma del components de la llista: {llista2} es igual = {sumallistes(llista2)}")


````

**3. Funció que pren com a paràmetre una llista de números i retorna la suma dels quadrats dels seus elements.**

```
def sumaquadrats ([4,3,2,1,8]) ---------> OUTPUT: 94
def sumaquadrats ([5,4,3,2,1]) ---------> OUTPUT: 55
```

````python
llista1 = [4, 3, 2, 1, 8]
llista2 = [5, 4, 3, 2, 1]

def sumaquadrats(llista):
    resultatquadrats = 0
    for pos in range(len(llista)):
        resultatquadrats = resultatquadrats + (llista[pos]*llista[pos])
    return resultatquadrats

print(f" La suma dels quadrats dels components de la llista {llista1} = {sumaquadrats(llista1)}")
print(f" La suma dels quadrats dels components de la llista {llista2} = {sumaquadrats(llista2)}")

````

````python
llista1=[4, 3, 2, 1, 8]
llista2=[5, 4, 3, 2, 1]

def parell(llistarevisar):
    llistaparells=[]
    for pos in range(len(llistarevisar)):
        if pos % 2 == 0:
            llistaparells.append(llistarevisar[pos])
    return(llistaparells)

print(f"La llista {llista1} té els seguents components en posicio parell: {parell(llista1)}")
print(f"La llista {llista2} té els seguents components en posicio parell: {parell(llista2)}")
````

**9. Funció que pren com a paràmetre una llista de números i retorna una altra llista amb els elements en posició senar.**

```
def senar ([4,3,2,1,8]) ---------> OUTPUT: [3,1]

```

````python
llista1=[4, 3, 2, 1, 8]
llista2=[5, 4, 3, 2, 1]

def senar(llistarevisar):
    llistasenars=[]
    for pos in range(len(llistarevisar)):
        if pos % 2 != 0:
            llistasenars.append(llistarevisar[pos])
    return(llistasenars)
print(f"La llista {llista1} té els seguents components en posicio imparell: {senar(llista1)}")
print(f"La llista {llista2} té els seguents components en posicio imparell: {senar(llista2)}")
````

**20. Funció que pren com a paràmetre una llista de cadenes i retorna una altra
llista amb les mateixes cadenes invertides.**

````python
llista1=["Bicicleta  ", "cullera", "poma "]
llista2=["Eat", "Pray ", "Love  ", "Hate"]

def inverteixcadenes(llistarevisar):
    llistainvertida = []
    for paraula in range(len(llistarevisar)):
        cadena = llistarevisar[paraula]
        cadenainvertida = cadena[::-1]
        llistainvertida.append(cadenainvertida)
    return(llistainvertida)

print(inverteixcadenes(llista1))
print(inverteixcadenes(llista2))
````
